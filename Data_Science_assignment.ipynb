<h1>Part 1: Data Cleaning</h1>

```python
cols = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
vals = np.arange(1,13)
```

```python
def data_cleaning():
    
    #To replace Null values with 0s
    df['How_many_Landed_on_our_Page?'] = df['How_many_Landed_on_our_Page?'].fillna(0)
    df['How_many_Landed_on_the_our_Page_and_clicked_on_a_button?'] = df['How_many_Landed_on_the_our_Page_and_clicked_on_a_button?'].fillna(0)
    
    #Replace Name of months with their respective number
    df['Month'] = df['Month'].replace(cols,vals)
    
    #replacing data in column 4(counting from 0)
    df['Where_Are_They_comming_from?'] = df['Where_Are_They_comming_from?'].replace(['Came_From_Google','Landed_on_the_page_Directly'],['Google','Direct_traffic'])
```

```python
#Calling data_cleaning() fuction to see changes
data_cleaning()
```

<h1>Part 2: Descriptive Statistics</h1>

```python
def descriptive_stats():
    
    #For Numeric Columns
    numerics = ['int64','float64']
    numeric_df = df.select_dtypes(include=numerics)
    print(numeric_df.shape)
    print(numeric_df.describe())
    
    print('*******************************************************************************************')
    
    #For non-numeric columns
    non_numeric = ['object']
    non_numeric_df = df.select_dtypes(include=non_numeric)
    print(non_numeric_df.shape)
    
    for col in non_numeric_df:
        print(non_numeric_df[col].unique())
        
    
```

```python
descriptive_stats()
```
